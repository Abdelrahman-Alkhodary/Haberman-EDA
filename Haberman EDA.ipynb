## Relevant Information
The dataset contains cases from a study that was conducted between 1958 and 1970 at the University of Chicago's Billings Hospital on the survival of patients who had undergone surgery for breast cancer.

Number of Instances: 306

Number of Attributes: 4 (including the class attribute)

### Attribute Information:

- Age of patient at time of operation (numerical)
- Patient's year of operation (year - 1900, numerical)
- Number of positive axillary nodes detected (numerical)
- Survival status (class attribute) 1 = the patient survived 5 years or longer 2 = the patient died within 5 year

you can download the dataset from https://www.kaggle.com/gilsousa/habermans-survival-data-set

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
haberman = pd.read_csv('haberman.csv',names=['age', 'op_year', 'axil_node', 'survival_status'])
```

```python
haberman.head()
```

```python
haberman['survival_status'].value_counts()
```

```python
haberman.describe()
```

```python
haberman[haberman['survival_status']==1].describe()
```

```python
haberman[haberman['survival_status']==2].describe()
```

```python
sns.pairplot(haberman, hue='survival_status', size=3)
```

```python
sns.FacetGrid(haberman, hue="survival_status", size=5) \
   .map(sns.distplot, "age") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(haberman, hue="survival_status", size=5) \
   .map(sns.distplot, "op_year") \
   .add_legend();
plt.show();
```

```python
sns.FacetGrid(haberman, hue="survival_status", size=5) \
   .map(sns.distplot, "axil_node") \
   .add_legend();
plt.show();
```

## Observation 
The Number of positive axillary nodes detected is better at indicating the survival status better than the ages of the patient and the operation year

```python
sns.boxplot(data=haberman, x='survival_status', y='age')
plt.show()
```

```python
sns.boxplot(data=haberman, x='survival_status', y='op_year')
plt.show()
```

```python
sns.boxplot(data=haberman, x='survival_status', y='axil_node')
plt.show()
```

### Observation
From the boxplot of the axil node we can see that patients with low number of axil node that have better chance to survive after 5 years of the operation
